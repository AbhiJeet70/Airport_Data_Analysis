```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
from torch_geometric.datasets import Airports

def analyze_and_visualize_airport_data(dataset_name, root="./datasets/Airport"):
    # Load the dataset
    dataset = Airports(root=root, name=dataset_name)
    data = dataset[0]

    print(f'Dataset: {dataset_name}')
    print('======================')
    print(f'Number of graphs: {len(dataset)}')
    print(f'Number of features: {dataset.num_features}')
    print(f'Number of classes: {dataset.num_classes}')
    print(data)
    print('===============================================================')

    # Get node features and the graph
    node_features = data.x.numpy()
    edge_index = data.edge_index.numpy()

    # Create a NetworkX graph from the edge index
    G = nx.Graph()
    edges = edge_index.T  # Transpose to get edges as pairs of nodes
    G.add_edges_from(edges)

    # Use KMeans to cluster the nodes based on their features
    num_clusters = dataset.num_classes  # Number of clusters is the number of classes
    kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(node_features)

    # Get the cluster labels for each node
    cluster_labels = kmeans.labels_

    # Dictionary to store nodes in each cluster
    cluster_nodes = {i: [] for i in range(num_clusters)}

    # Assign nodes to their respective clusters
    for idx, label in enumerate(cluster_labels):
        cluster_nodes[label].append(idx)

    # Find the centroids of each cluster
    cluster_centroids = kmeans.cluster_centers_

    # Function to find nodes closest to each centroid
    closest_nodes = {}
    for cluster in range(num_clusters):
        closest, _ = pairwise_distances_argmin_min(cluster_centroids[cluster].reshape(1, -1), node_features)
        closest_nodes[cluster] = closest[0]  # Ensure only one closest node is selected

    # Store information about clusters and selected nodes
    cluster_info = {}

    # Calculate number of nodes in each cluster and details of selected nodes
    for cluster in range(num_clusters):
        num_nodes = len(cluster_nodes[cluster])
        selected_node = closest_nodes[cluster]
        selected_node_degree = G.degree[selected_node]
        selected_node_features = node_features[selected_node]

        cluster_info[cluster] = {
            'num_nodes': num_nodes,
            'selected_node': selected_node,
            'selected_node_degree': selected_node_degree,
            'selected_node_features': selected_node_features
        }

    # Visualization using force-directed layout
    plt.figure(figsize=(12, 8))
    plt.title(f'Graph Visualization with Clusters and Closest Nodes to Centroids for {dataset_name}')

    # Define colors for clusters
    colors = plt.cm.rainbow(np.linspace(0, 1, num_clusters))

    # Draw nodes without edges for separation
    pos = nx.spring_layout(G)
    nx.draw_networkx_nodes(G, pos, node_size=0, alpha=0)

    # Draw nodes with different colors for each cluster and no edges
    for cluster, color in zip(cluster_nodes.keys(), colors):
        nx.draw_networkx_nodes(G, pos, nodelist=cluster_nodes[cluster],
                               node_color=color, node_size=200, alpha=1.0,
                               label=f'Cluster {cluster}')

    # Highlight nodes closest to centroids in a different color for each cluster
    for cluster, node in closest_nodes.items():
        color = colors[cluster]
        nx.draw_networkx_nodes(G, pos, nodelist=[node],
                               node_color=color, node_size=300, alpha=1.0, node_shape='s',
                               linewidths=1.0, edgecolors='k')

    # Add legend and show plot
    plt.legend()

    # Print cluster information
    print(f'Cluster Information for {dataset_name}:')
    for cluster, info in cluster_info.items():
        print(f'Cluster {cluster}:')
        print(f'  Number of nodes: {info["num_nodes"]}')
        print(f'  Selected node: {info["selected_node"]}')
        print(f'  Degree of selected node: {info["selected_node_degree"]}')
        print(f'  Features of selected node: {info["selected_node_features"]}')
        print('-------------------------')

    plt.show()

# Dataset names
dataset_names = ["USA", "Brazil", "Europe"]

# Loop over each dataset name and analyze/visualize
for name in dataset_names:
    analyze_and_visualize_airport_data(name)
```
